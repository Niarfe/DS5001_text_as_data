```python
import pandas as pd
```

```python
epub = open("pg2701.txt", 'r', encoding='utf-8-sig').readlines()
df = pd.DataFrame(epub, columns=['line_str'])
df.index.name = 'line_num'
df.line_str = df.line_str.str.strip()
```

```python
title = df.loc[0].line_str.replace('The Project Gutenberg eBook of ', '')
df['title'] = title
print(title)
```

```python
df.head(10)
```

```python
# How many lines does the file have
print(f"The number of lines in the file is {len(df)}")
```

```python
a = df.line_str.str.match(r"\*\*\*\s*START OF (THE|THIS) PROJECT")
b = df.line_str.str.match(r"\*\*\*\s*END OF (THE|THIS) PROJECT")
an = df.loc[a].index[0]
bn = df.loc[b].index[0]
df = df.loc[an + 1 : bn - 2]
df
```

```python
# When does the content begin
first_row_past_header = df.iloc[0]
print(f"The first row past the Gutenberg standard header is {first_row_past_header.name}")
```

```python
df_non_empty = df[df['line_str'] != '']
first_content_row_past_header = df_non_empty.iloc[0]
print(f"The first NON-EMPTY row past the Gutenberg standard header is {first_content_row_past_header.name}")
```

```python
# When does the content end
first_row_past_header = df.iloc[-1]
print(f"The first row past the Gutenberg standard header is {first_row_past_header.name}")
```

```python
last_content_row_before_footer = df_non_empty.iloc[-1]
print(f"The first NON-EMPTY row before the Gutenberg standard footer is {last_content_row_before_footer.name}")
```
